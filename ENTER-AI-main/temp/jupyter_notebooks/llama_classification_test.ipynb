```python
data = pd.read_csv('./data.csv')['comment']
```

```python
model_path = "mistralai/Mistral-7B-Instruct-v0.2"
    

comment_list = []
result_list = []
end_list = []

total_start = time.time()

model_name = model_path.split('/')[-1]
lp = LangchainPipline(user_id='asdf1234', model_path=model_path)
    

for comment in data:
    start = time.time()
    result = lp.chain(comment)
    print(result)
    end = time.time() - start
    end_list.append(end)
    comment_list.append(comment)
    result_list.append(result)

total_end = time.time() - total_start

pd.DataFrame({'comment':comment_list,'result':result_list,'end_time':end_list}).to_csv(f'./{model_name}_{total_end}.csv',index=False)
```
